```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## sorting the data based on the population in 20222
# df.head()
df.sort_values(by='2022 Population', ascending=True).head()
```

```python
## the correlation between features(columns)
df.corr()
```

```python
## Visualize the correlation between features
sns.heatmap(df.corr(), annot=True)

plt.rcParams['figure.figsize'] = (20, 8)

plt.show()
```

```python
## average of each column per continent

# df.columns
df.groupby('Continent').mean().sort_values(by='2022 Population', ascending=True)
```

```python
pop_df = df.groupby('Continent')[['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2015 Population','2020 Population', '2022 Population']].mean().sort_values(by='2022 Population', ascending=True)
```

```python
## plotting the change in population in continets over time
pop_df1 = pop_df.transpose()
pop_df1.plot()
```
